```python
# prompt: convert death column datatype to int

import pandas as pd
df['Death'] = pd.to_numeric(df['Death'], errors='coerce').astype('Int64')
```

```python
# This helps in tracking the number of currently infected people.
df["Active_Cases"] = df["Total Confirmed cases"] - (df["Death"] + df["Cured/Discharged/Migrated"])
```

```python
#Time series plot or line plot over time
import matplotlib.pyplot as plt
import pandas as pd

# Ensure Date is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Grouping and plotting
plt.figure(figsize=(12, 6))
df_grouped = df.groupby("Date")["Total Confirmed cases"].sum()
df_grouped.plot(kind="line", marker="o", color="blue")

plt.title("Total Confirmed COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_grouped = df.groupby("Date")["Active_Cases"].sum()
df_grouped.plot(kind="line", marker="o", color="red")

plt.title("Active COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Active Cases")
plt.grid()
plt.show()
```

```python
# import seaborn as sns
# plt.figure(figsize=(12, 6))
# df_latest = df[df["Date"] == df["Date"].max()]  # Get latest date's data
# df_latest = df_latest.sort_values(by="Total Confirmed cases", ascending=False)

# sns.barplot(x="Total Confirmed cases", y="Name of State / UT", data=df_latest, palette="coolwarm")
# plt.title("State-wise Total COVID-19 Cases")
# plt.xlabel("Total Cases")
# plt.ylabel("State/UT")
# plt.show()
# import plotly.express as px
# import pandas as pd

# df_latest = df[df["Date"] == df["Date"].max()]
# df_latest = df_latest.sort_values(by="Total Confirmed cases", ascending=False)

# # Calculate total cases for percentage
# total_cases = df_latest["Total Confirmed cases"].sum()
# df_latest["Percentage"] = (df_latest["Total Confirmed cases"] / total_cases) * 100

# fig = px.bar(
#     df_latest,
#     x="Total Confirmed cases",
#     y="Name of State / UT",
#     orientation='h',
#     color="Total Confirmed cases",
#     color_continuous_scale="thermal",
#     text=df_latest["Percentage"].apply(lambda x: f'{x:.2f}%')
# )

# fig.update_layout(
#     title="State-wise Total COVID-19 Cases (Hover to see %)",
#     xaxis_title="Total Cases",
#     yaxis_title="State/UT",
# )

# fig.show()

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(12, 6))

# Get latest date's data
df_latest = df[df["Date"] == df["Date"].max()]
df_latest = df_latest.sort_values(by="Total Confirmed cases", ascending=False)

# Calculate total cases
total_cases = df_latest["Total Confirmed cases"].sum()

# Plot barplot
sns.barplot(x="Total Confirmed cases", y="Name of State / UT", data=df_latest, palette="coolwarm")

# Add % text on bars
for index, value in enumerate(df_latest["Total Confirmed cases"]):
    percentage = (value / total_cases) * 100
    plt.text(value + 1000,  # Shift text slightly right
             index,
             f'{percentage:.2f}%',
             va='center',
             fontsize=10,
             color='black')

plt.title("State-wise Total COVID-19 Cases with % Share")
plt.xlabel("Total Cases")
plt.ylabel("State/UT")
plt.tight_layout()
plt.show()

```

```python
df_latest = df[df["Date"] == df["Date"].max()]  # Latest data

labels = ["Recovered", "Deaths", "Active Cases"]
sizes = [df_latest["Cured/Discharged/Migrated"].sum(), df_latest["Death"].sum(), df_latest["Active_Cases"].sum()]
colors = ["green", "red", "orange"]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=140)
plt.title("Recovery vs Deaths vs Active Cases")
plt.legend()
plt.show()
```

```python
#Time Series Heatmap
df_pivot = df.pivot_table(values="Total Confirmed cases", index="Name of State / UT", columns="Date")
plt.figure(figsize=(12, 8))
sns.heatmap(df_pivot, cmap="Reds", linewidths=0.5)
plt.title("COVID-19 Spread Across States Over Time")
plt.xlabel("Date")
plt.ylabel("State/UT")
plt.show()
```

```python
# Geospatial Scatter Plot with more hover data
import plotly.express as px

fig = px.scatter_geo(
    df_latest,
    lat="Latitude",
    lon="Longitude",
    size="Total Confirmed cases",
    color="Total Confirmed cases",
    hover_name="Name of State / UT",
    hover_data={
        "Total Confirmed cases": True,
        "Active_Cases": True,
        "Death": True,
        "Latitude": False,  # hide lat/lon from hover
        "Longitude": False
    },
    projection="orthographic",
    title=" COVID-19 Spread Across India (Confirmed, Active, Deaths)"
)

fig.show()
```

```python
#word cloud
from wordcloud import WordCloud

state_cases = dict(zip(df_latest['Name of State / UT'], df_latest['Total Confirmed cases']))

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(state_cases)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```

```python
#Line chart
df['Daily_Growth_Rate'] = df.groupby('Name of State / UT')["Total Confirmed cases"].pct_change() * 100
# pct_change=((Today - Yesterday) / Yesterday) * 1


df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Plotting Daily Growth Rate
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Date', y='Daily_Growth_Rate', marker='o', color='orange')

plt.title('COVID-19 Daily Growth Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Bubble Chart
import plotly.express as px
latest = df.sort_values("Date").groupby("Name of State / UT").last().reset_index()
latest["Recovery Rate"] = latest["Cured/Discharged/Migrated"] / latest["Total Confirmed cases"]
latest["Death Rate"] = latest["Death"] / latest["Total Confirmed cases"]

top10 = latest.sort_values("Total Confirmed cases", ascending=False).head(10)

fig = px.scatter(top10,
                 x="Recovery Rate",
                 y="Death Rate",
                 size="Total Confirmed cases",
                 color="Name of State / UT",
                 hover_name="Name of State / UT",
                 title=" Top 10 States: Recovery vs Death Rate",
                 size_max=60)

fig.show()
```

```python

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Group by State and Month
monthly_df = df.groupby(
    ['Name of State / UT', pd.Grouper(key='Date', freq='M')]
)[['Active_Cases', 'Cured/Discharged/Migrated']].sum().reset_index()
```

```python
#Multi-Line Plot (or Line Chart with Multiple Series)
fig1 = px.line(
    monthly_df,
    x='Date',
    y='Active_Cases',
    color='Name of State / UT',
    title='Month-wise Active COVID-19 Cases in States',
    markers=True
)

fig1.update_layout(
    xaxis_title='Month',
    yaxis_title='Active Cases',
    legend_title='States/UTs'
)

fig1.show()
```

```python
#Multi-Line Plot (or Line Chart with Multiple Series)
fig2 = px.line(
    monthly_df,
    x='Date',
    y='Cured/Discharged/Migrated',
    color='Name of State / UT',
    title='Month-wise Discharged COVID-19 Cases in States',
    markers=True
)

fig2.update_layout(
    xaxis_title='Month',
    yaxis_title='Discharged Cases',
    legend_title='States/UTs'
)

fig2.show()
```

```python
# Treemap
import pandas as pd
import plotly.express as px

# Ensure columns are numeric
df["Total Confirmed cases"] = pd.to_numeric(df["Total Confirmed cases"], errors="coerce")
df["Death"] = pd.to_numeric(df["Death"], errors="coerce")
df["Cured/Discharged/Migrated"] = pd.to_numeric(df["Cured/Discharged/Migrated"], errors="coerce")
df["New cases"] = pd.to_numeric(df["New cases"], errors="coerce")
df["New recovered"] = pd.to_numeric(df["New recovered"], errors="coerce")

# Take the latest data
latest = df.sort_values("Date").groupby("Name of State / UT").last().reset_index()

# Build the treemap
fig = px.treemap(
    latest,
    path=[px.Constant("India"), "Name of State / UT"],
    values="Total Confirmed cases",
    color="Death",
    color_continuous_scale="Reds",
    hover_data={
        "Total Confirmed cases": True,
        "Death": True,
        "Cured/Discharged/Migrated": True,
        "New cases": True,
        "New recovered": True
    },
    title=" COVID-19 Treemap: Indian States by Confirmed Cases & Deaths"
)

# Show only state names inside the boxes (no numbers)
fig.update_traces(textinfo="label", selector=dict(type='treemap'))

# Final layout tweaks
fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

fig.show()
```
